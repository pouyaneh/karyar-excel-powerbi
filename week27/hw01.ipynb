```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initial money
money = 10000

# number of simulations
n_games = 1_000_000_000

step = 10
```

```python
# simulate dice rolls: random integers between 1 and 6
dice_rolls = np.random.randint(1, 7, size=n_games)

# win condition: roll == 6
wins = dice_rolls == 6
```

```python
# calculate net gains/losses
# each play costs $10
# win: earn $60 - $10 = +$50 net
# loss: lose $10
net_gains = np.where(wins, 50, -10)
```

```python
money_progress = [money]
for i in range(0, n_games, step):
    money += np.sum(net_gains[i:i+step])
    money_progress.append(money)
```

```python
games_played = np.arange(0, n_games + 1, step)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(games_played, money_progress, color='blue', linewidth=1.5)
plt.title("Remaining Money Over Time While Playing the Dice Game")
plt.xlabel("Number of Games Played")
plt.ylabel("Remaining Money ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
